```python
# import packages
import numpy as np
import pandas as pd
```

```python
# get the data by reading the csv file
# select only the columns needed
df = pd.read_csv('^GSPC-2000-01-01-2017-06-30.csv', index_col=0)
df = df[['Close']]
df.info()
```

```python
# plot the closing prices
df['Close'].plot(grid=True, figsize=(12,6))
```

```python
# add 42 and 252 moving averages
df['42d'] = np.round(df['Close'].rolling(window=42).mean(), 2)
df['252d'] = np.round(df['Close'].rolling(window=252).mean(), 2)

# print the tail
df.tail()
```

```python
# plot the closing prices and the newly added moving averages
df.plot(grid=True, figsize=(12,6))
```

```python
# generate traind signals
# long  - 42d trend is SD points above the 252d trend
# cash  - 42d trend within a range of +/- SD points around 252d trend
# short - 42d trend is SD pints below the 252d trend

# set SD to 50(can be optimized later)
SD = 50

# new column for different between 42d and 252d
df['42-252'] = df['42d'] - df['252d']

# Regime = 1, buy 
# Regime = -1, sell
# Regime = 0, hold in cash
df['Regime'] = np.where(df['42-252'] > SD, 1, 0)
df['Regime'] = np.where(df['42-252'] < -SD, -1, df['Regime'])

# print the number of rows with the values
df['Regime'].value_counts()
```

```python
# Base on the calue counts,
# On 2073 days, we will be long, 
# On 950 days we will be short 
# On 1378 days we will stay in cash

# plot the regime
df['Regime'].plot(lw=1.5, figsize=(12,6))
```

```python
# calculate returns
# calculate the returns of the market
df['Market'] = np.log(df['Close'] / df['Close'].shift(1))

# calculate the returns of the strategy
df['Strategy'] = df['Regime'].shift(1) * df['Market']

# plot the cumulative returns of Market and Strategy
df[['Market', 'Strategy']].cumsum().apply(np.exp).plot(grid=True, figsize=(12,6))
```
